```python
imSize = 1946
```

```python
# prepare coordinates of points

myRadList = [30,50,100,150,200,250,500,1000]

for im_num in range(425):

    coords = np.array([df.loc[df.t==im_num,'centroid-0'],
                         df.loc[df.t==im_num,'centroid-1']]).T
    print(f'shape of the cell profiler array: {coords.shape}')

    # calculate distances between points
    myDist = distance.cdist(coords, coords, 'euclidean')
    print(f'shape of the distance  array: {myDist.shape}')

    for myRad in myRadList:
    
        # find cells within the radius
        cellNum=np.sum(myDist<myRad,axis=0)-1
        print(f'shape of the density measure array: {cellNum.shape}')

        df.loc[df.t==im_num,f'density_rad_{myRad}'] = cellNum
        
        # remove info for cells that are too close to the edge
        edgeCells = [(df['centroid-0']<myRad) | (df['centroid-1']<myRad) | 
                    (df['centroid-0']>(imSize-myRad)) | (df['centroid-1']>(imSize-myRad))]
        
        df.loc[(np.array(df.t==im_num) & np.array(edgeCells).squeeze()),f'density_rad_{myRad}'] = np.NaN
        
        print(myRad)
        print(np.sum((np.array(df.t==im_num) & np.array(edgeCells).squeeze())))
```
